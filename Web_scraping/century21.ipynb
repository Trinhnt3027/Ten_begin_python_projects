```python
#import packages

import requests
from bs4 import BeautifulSoup
import pandas
```

```python
# Get total page

r = requests.get("https://pythonizing.github.io/data/real-estate/rock-springs-wy/LCWYROCKSPRINGS/t=0&s=0.html")
content = r.content
soup = BeautifulSoup(content)
numberPages = int(soup.find_all("a", {"class":"Page"})[-1].text)
```

```python
#get data

base_url = "https://pythonizing.github.io/data/real-estate/rock-springs-wy/LCWYROCKSPRINGS/t=0&s={}.html"
listItems = []
    
for pageIndex in range(numberPages):
    page_url = base_url.format(pageIndex*10)
    r = requests.get(page_url)
    content = r.content
    soup = BeautifulSoup(content)
    
    items = soup.find_all("div", {"class":"CenterLaneCard propertyCard"})

    for item in items:
        dictItem = {}
        dictItem['price'] = item.find("h4", {"class":"propPrice"}).text.strip()
        infoDetail = item.find("div", {"class":"primaryDetails"})
        addressInfo = infoDetail.find_all("span", {"class":"propAddressCollapse"})
        dictItem['address'] = addressInfo[0].text
        dictItem['locality'] = addressInfo[1].text

        listItems.append(dictItem)
        
df = pandas.DataFrame(listItems)
df.to_csv("Output.csv")
df
```
